```python
# demonstration of the central limit theorem
import numpy as np
from numpy.random import randint
from numpy import mean
from matplotlib import pyplot


pmeans = [mean(np.random.normal(0,10, 1000)) for _ in range(100)]
pmeans2 = [mean(np.random.normal(0,10, 10000)) for _ in range(100)]
print(np.std(pmeans,ddof=1))
print(np.std(pmeans2,ddof=1))
# plot the distribution of sample means
pyplot.subplot(1,2,1)
pyplot.hist(pmeans)
pyplot.subplot(1,2,2)
pyplot.hist(pmeans2)
pyplot.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import ttest_ind

sonuç = []
sonuç2 = []
tt_test=[]
normalpara = np.random.binomial(1, 0.5, 10000)
for atış_sayısı in range(1,1000):
    atış = np.random.binomial(1, 0.75, atış_sayısı)
    atışların_ortalaması = atış.mean()
    sonuç.append(atışların_ortalaması)
    atış2 = np.random.binomial(1, 0.5, atış_sayısı)
    atışların_ortalaması2 = atış2.mean()
    sonuç2.append(atışların_ortalaması2)
    tt_test.append(ttest_ind(normalpara, atış, equal_var=False).pvalue)

df = pd.DataFrame({ 'atış' : sonuç})
df = pd.DataFrame({ 'atış2' : sonuç2})
plt.subplot(1,2,1)
plt.plot(sonuç)
plt.plot(sonuç2)
plt.subplot(1,2,2)
plt.plot(tt_test[1:50])
plt.show()
```
